## Brute search in the text files

```python
from lxml import etree as et
def arg_in_actors(actor_sentence: str, pronoun=True, actor_xml_path="data/info/actors.xml") :
    """
    Returns a list with all valid actors of the sentence that can be found in actors.xml
    Parameters
    ----------
    pronoun : bool
        False, if we don't want to query pronouns in the xml file, True if we want to return pronons as actors
    actor_sentence : str
        The ARG0 argument of AllenNLP
    actor_xml_path : str
        Path to actors.xml

    Returns
    -------
    list
        A list with all the actors present in ARG0
    """
    actor_sentence = actor_sentence.lower()
    lst_actors = []
    # Parsing the xml actors.xml
    tree = et.parse(actor_xml_path)
    root = tree.getroot()
    # 4 type of actors : country observer, generic others
    for type_actor in root:
        for actor in type_actor:
            # Excluding generic actors (committee /delegate /party / chair /manager )
            if actor.tag != "generic":
                # if pronoun == False, we skip the pronouns from the xml file
                if pronoun or (type_actor.tag != "pronoun"):
                    # Regex to match the noun phrase to the xml.
                    reg = r'\b' + re.escape(actor.text) + r'\b'
                    if re.search(reg.lower(), actor_sentence.lower(), re.IGNORECASE):
                        # Special case for Guinea / NOT PRETTY BUT MOSTLY IRRELEVANT
                        if actor.text == "Guinea":
                            if re.search(r"papua|equatorial|bissau", actor_sentence.lower()) and (
                                    actor_sentence.lower().count("guinea") == 1):
                                pass
                            else:
                                lst_actors.append(actor.attrib["dbpedia"])
                        else:
                            # If there's a dbpedia name, we use it
                            if "dbpedia" in actor.attrib:
                                lst_actors.append(actor.attrib["dbpedia"])
                            else:
                                lst_actors.append(actor.text)
    lst_actors = [actor.replace("_", " ") for actor in lst_actors]
    return list(set(lst_actors))
```

```python
import glob
import re

import spacy
# !python -m spacy download en_core_web_trf
nlp = spacy.load("en_core_web_trf")
nlp.disable_pipes('transformer', 'attribute_ruler', 'ner')

enb = glob.glob("/home/kevin-main/PycharmProjects/Relation_extraction_ENB/data/raw/txt_boiler/*")
tree = et.parse("/home/kevin-main/PycharmProjects/Relation_extraction_ENB/data/info/enb.xml")
res_list = []
for file in enb:
    # We cross the name of the file with the enb_co_no_summary.xml to acquire the year of the conference
    num = file.split("txt_boiler/")[-1].split(".txt")[0]
    year = int(tree.xpath("//event[urls/url/@num='" + num + "']/date/@year")[0])
    event = tree.xpath("//event[urls/url/@num='" + num + "']/@title")[0]
    # Reading the file
    with open(file, "r", encoding="utf-8") as f:
        for paragraph in f:
            # Using spaCy to split paragraph in sentences.
            doc = nlp(paragraph.rstrip())
            for sentence in doc.sents:
                # List of actors in the sentence
                lst_actors_in_sentence = arg_in_actors(sentence.text, pronoun=False, actor_xml_path="/home/kevin-main/PycharmProjects/Relation_extraction_ENB/data/info/actors.xml")
                # If at least one non_eu actors is the list of actors and at least one EU actors is in the list of actors, then proceed
                if any(country in lst_actors_in_sentence for country in lst_non_eu_countries) and any(country in lst_actors_in_sentence for country in lst_eu_countries):
                    lst_non_eu_countries_in_sentence = list(set(lst_actors_in_sentence) & set(lst_non_eu_countries))
                    lst_eu_countries_in_sentence = list(set(lst_actors_in_sentence) & set(lst_eu_countries))
                    lst = [year, event, num, ", ".join(lst_non_eu_countries_in_sentence), sentence.text]
                    res_list.append(lst)



df_res = pd.DataFrame(res_list, columns=['Year', 'Event', 'Num', 'Non_EU_country', 'Sentence'])
df_res
```

```python
df_res = df_res.sort_values(by='Year')
df_res.to_excel("full_enb_results_year.xlsx")
```

```python
#  Excel 1 country = 1 spreadsheet
with pd.ExcelWriter("full_enb_results_by_country.xlsx") as writer:
    for country in lst_non_eu_countries:
        cond_on_country = df_res['Non_EU_country'].str.contains(country)
        df_country = df_res[cond_on_country]
        df_country['Non_EU_country'] = country
        df_country = df_country.sort_values(by='Year')
        df_country.to_excel(writer, sheet_name=country)
```
